# Prepare simulation data

Run this notebook to preprocess the raw simulation data in `Simulations.jld2` into a form suitable for training of neural functionals.

```julia
using JLD2
include("utils.jl")
```

```julia
c1_profiles = load("Simulations.jld2", "c1_profiles")
ρ_profiles = load("Simulations.jld2", "ρ_profiles")
ϕ_profiles = load("Simulations.jld2", "ϕ_profiles");
```

```julia
ρ_windows, c1_values, ϕ_functions = generate_inout(ρ_profiles, c1_profiles, ϕ_profiles; window_width=2.0, dx=0.01);
```

```julia
using NPZ 
metainput = vcat(ρ_windows, ϕ_functions,c1_values);
```

```julia
npzwrite("metainput.npy", metainput)
```
